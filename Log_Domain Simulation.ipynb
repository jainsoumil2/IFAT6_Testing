$$
\require{cancel}
$$

For a single synapse and compartment:

<img src="Synapse.png" width="615" height="130">

$$ C \frac{dv_{s}}{dt} = I_o \frac{W}{L} e^{ \frac{\kappa v_{gb}}{V_{th}} } (e^{ -\frac{v_{sb}}{V_{th}} } - e^{ -\frac{v_{db}}{V_{th}} }) $$

Setting $v_d = v_{rev}$ and $v_s = v_m$

$$ C \frac{dv_m}{dt} = I_o \frac{W}{L} e^{ \frac{\kappa v_g}{V_{th}} } (e^{ -\frac{v_m}{V_{th}} } - e^{ -\frac{v_{rev}}{V_{th}} }) \tag{1} $$ 

where $V_{th}$ is the thermal voltage.

In the **log-domain**:

$$ C^* \frac{dV_{m}^{*}}{dt} = G^* (V_{rev}^{*} - V_{m}^{*}) \tag{2} $$

where the pseudo-parameters of conductance 
$G^* = \frac{1}{V_{th}} I_o \frac{W}{L} e^{ \frac{\kappa v_g}{v_{th}} }$, reversal potential 
$V_{rev}^{*} = -V_{th} e^{ -\frac{v_{rev}}{V_{th}} } $ and membrane potential 
$V_{m}^{*} = -V_{th} e^{ -\frac{v_m}{V_{th}} } $.

To make equations (1) and (2) equivalent, 
$$ C^* \frac{dV_{m}^{*}}{dt} = C \frac{dv_m}{dt} \tag{3} $$

where $ C^{*} \frac{dV_{m}^{*}}{dt} $ can be evaluated as

$$ C^{*} \frac{dV_{m}^{*}}{dt} = C^*. -\cancel{V_{th}}. e^{ -\frac{v_m}{V_{th}} }. -\frac{1}{\cancel{V_{th}}}. \frac{dv_m}{dt} = C^*. e^{ -\frac{v_m}{V_{th}} }. \frac{dv_m}{dt} $$

Substitute back in (3)

$$ C^*. e^{ -\frac{v_m}{V_{th}} }. \cancel{\frac{dv_m}{dt}} = C \cancel{\frac{dv_m}{dt}} $$

gives 

$$ C^* = C e^{ \frac{v_m}{V_{th}} } $$

Multiply and divide by $-V_{th}$ to get

$$ C^* = \frac{-V_{th}. C}{(-V_{th}.e^{ -\frac{v_m}{V_{th}} })} = -\frac{V_{th}}{V_m^*}. C $$

Substitute back in (2) and extend the equation to multiple synapses and compartments:

$$ -\frac{V_{th}}{V_{m_{i}}^*} C \frac{dV_{m_{i}}^{*}}{dt} = \sum_{j}G_{ij}^* (V_{rev_{ij}}^{*} - V_{m_{i}}^{*}) $$

where 

\begin{align}
V_{m_{i}}^{*} &= -V_{th} e^{ -\frac{v_{m_{i}}}{V_{th}} } \\
V_{m_{i}}^{*} &\leq 0 \\
V_{rev_{ij}} &= -V_{th} e^{ -\frac{v_{rev_{ij}}}{V_{th}} } \\
G_{ij}^* &= \frac{1}{V_{th}} I_o \frac{W}{L} e^{ \frac{\kappa v_{g_{ij}}}{V_{th}} } 
\end{align}

$$ V_{m_{i}}^* > \theta_{i}^* = -V_{th} e^{-\frac{\theta_i}{V_{th}}} => Fire $$

The last equation for the threshold of firing in the log-domain can be derived from the original equations as shown below.

Compared to the **original equation**:

$$ C \frac{dv_{m_{i}}}{dt} = \sum_{j} I_o \frac{W}{L} e^{ \frac{\kappa v_{g_{ij}}}{V_{th}} } (e^{ -\frac{v_{m_{i}}}{V_{th}} } - e^{ -\frac{v_{rev_{ij}}}{V_{th}} }) $$ 

$$ v_{m_{i}} > \theta_{i} => Fire $$

which implies

\begin{align}
-V_{th} e^{ -\frac{v_{m_{i}}}{V_{th}} } &> -V_{th} e^{ -\frac{\theta_{i}}{V_{th}} } \\
V_{m_{i}}^{*} &> -V_{th} e^{ -\frac{\theta_{i}}{V_{th}} } 
\end{align}

because $ -e^{ -\frac{v_{m_{i}}}{V_{th}} } $ is a monotonically increasing function of $ v_{m_{i}} $. As $v_{m_i}$ increases, the fraction $\frac{v_{m_i}}{V_{th}}$ increases, making the negative exponent less negative, which in turn makes the entire expression larger (less negative).


**Weights:**

$$ G_{ij}^* = \frac{1}{V_{th}} I_o \frac{W}{L} e^{ \frac{\kappa v_{g_{ij}}}{V_{th}} } $$

$$ \Delta G_{ij}^* = \frac{1}{V_{th}} I_o \frac{W}{L} \Delta e^{ \frac{\kappa v_{g_{ij}}}{V_{th}} } = \frac{1}{V_{th}} I_o \frac{W}{L} \frac{\kappa}{V_{th}} e^{ \frac{\kappa v_{g_{ij}}}{V_{th}} } \Delta v_{g_{ij}} $$

<img src="Synapse_impulse_input.png" width="218" height="183">

\begin{align}
\Delta v_{g_{ij}} &= \Delta t \frac{I_s}{C_{syn}} \\
&= \Delta t \frac{ I_{o_p} \frac{W_p}{L_p} e^{\frac{\kappa (v_{dd} - v_{g_{ij}})}{V_{th}}} e^{ \frac{-(v_{dd} - v_s)}{V_{th}} } }{C_{syn}} \\
&= \Delta t \frac{ I_{o_p} \frac{W_p}{L_p} e^{ \frac{\kappa v_{dd} - v_{dd}}{V_{th}} } e^{ \frac{-\kappa v_{g_{ij}}}{V_{th}} } e^{ \frac{v_s}{V_{th}} } }{C_{syn}} \\
&= \Delta t \frac{ I_{o_p} e^{\frac{-(1 - \kappa)v_{dd}}{V_{th}}} \frac{W_p}{L_p} e^{ \frac{-\kappa v_{g_{ij}}}{V_{th}} } e^{ \frac{v_s}{V_{th}} } }{C_{syn}} \\
&= \Delta t \frac{ I_{o_p} e^{\frac{-(1 - \kappa)v_{dd}}{V_{th}}} \frac{W_p}{L_p} e^{ \frac{-\kappa v_{g_{ij}}}{V_{th}} } e^{ \frac{v_s}{V_{th}} } }{C_{syn}} \\
\\
\text{=> } \Delta G_{ij}^* &= \frac{1}{V_{th}} I_{o_n} \frac{W_n}{L_n} \frac{\kappa}{V_{th}} \cancel{e^{ \frac{\kappa v_{g_{ij}}}{V_{th}} }} \frac{ \Delta t I_{o_p} e^{\frac{-(1 - \kappa)v_{dd}}{V_{th}}} \frac{W_p}{L_p} \cancel{e^{ \frac{-\kappa v_{g_{ij}}}{V_{th}} }} e^{ \frac{v_s}{V_{th}} }  }{C_{syn}} \\
&= \frac{ \kappa I_{o_n} I_{o_p} \frac{W_n}{L_n} \frac{W_p}{L_p} e^{\frac{-(1 - \kappa)v_{dd}}{V_{th}}} }{ V_{th}^2 C_{syn} } e^{\frac{v_s}{V_{th}}} \Delta t
\end{align}


### Subthreshold Channel Current in Saturation
### For an nch25_dnw Device used as a Synapse Element
### Estimate $\kappa / V_{th}$ parameter for the subthreshold channel current in saturation expressed as $ I = I_o e^{ \frac{\kappa v_g}{V_{th}} } $ where the source and body terminal are shorted to the lowest power supply, ground.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

%matplotlib qt

plt.rcParams['text.usetex'] = True
```

```python
df = pd.read_excel('Synapse_Device_Characterisation.xlsx', 
                   sheet_name = 'nch25_dnw subvt saturation',
                   usecols = [0,1,2,3,4,5,6],
                   skiprows = 3,
                   nrows = 501,
                   header = None)

Vg = df[0].values   # Convert to a NumPy array using .values, which is more efficient for iterating and performing calculations.
I = df[1].values
I_vs_20m = df[2].values
I_vs_40m = df[3].values
I_vs_60m = df[4].values
I_vs_80m = df[5].values
I_vs_100m = df[6].values
```

```python
dI_dVg = np.gradient(I, Vg)  # Calculate derivative

plt.figure(figsize=(10, 6))
plt.plot(Vg, I, label='I-Vg Curve')
plt.plot(Vg, dI_dVg, '--', label='Derivative of I-Vg')
plt.yscale('log')  # Use log scale for current
plt.xlabel('Vg (V)')
plt.ylabel('I (A)')
plt.grid(which='both', axis='both', linestyle=':', alpha=0.5)
plt.legend()
plt.title('Subthreshold Channel Current in Saturation', fontsize=14)
```

```python
# Calculate dI/dVg normalized by current
normalized_transconductance = dI_dVg / I  # Normalize by current

# Plot normalized transconductance
plt.figure(figsize=(15, 6))
plt.plot(I, normalized_transconductance, 'r-', label='(dI/dVg)/I')
plt.axhline(y=27, color="black", linestyle=":", alpha=0.5, label='(dI/dVg)/I = 27')

# Add annotations for data points
for i, (x, y) in enumerate((zip(I, normalized_transconductance))):
    if x < 1e-5 and i%5==0:
        # Add the marker
        plt.plot(x, y, 'ro', ms=5)  # 'ro' creates a red circle marker, ms=5 sets marker size
        
        # Add the annotation
        label = "{:.2f}".format(y)
        plt.annotate(label,
                    (x, y),  # Point being annotated
                    xytext=(0, 5),  # Offset for text position
                    textcoords="offset points",  # Use offset coordinates
                    ha='left',  # Horizontal alignment
                    rotation=45)  # Rotate text

plt.xscale('log')  # Use log scale for current
plt.xlabel('I (A)')
plt.ylabel('(dI/dVg)/I (V⁻¹)')
plt.grid(which='both', axis='both', linestyle=':', alpha=0.5)
plt.legend(framealpha=0.2)
plt.title('$g_m/I_D$ versus $I_D$', pad='25', fontsize=14)
plt.xlim([1e-11, 1e-6])
plt.ylim([10, 30])
```

```python
# Select region where current is between specific values
exp_region_indices = np.where((I > 1e-10) & (I < 1e-8))
Vg_exp = Vg[exp_region_indices]
I_exp = I[exp_region_indices]
```

```python
# Define exponential function
def exp_func(Vg, I0, slope):
    return I0 * np.exp(slope * Vg)

# Fit the curve
popt, pcov = curve_fit(exp_func, Vg_exp, I_exp, p0=(5.1172e-12, 27.0128))
I_fit = exp_func(Vg, *popt)

print(f'I0: {popt[0]:.4e}, kappa/Vth: {popt[1]:.4f}')
#print(np.linalg.cond(pcov))
#print(np.diag(pcov))
```

```python
plt.figure(figsize = (10, 6))
plt.plot(Vg, I,
         label='I-Vg Curve',
         linestyle='', 
         marker='o',
         markerfacecolor='none',
         markeredgewidth=0.5)
plt.plot(Vg, I_fit, 'k-', 
         label='I-Vg Fitted Curve', 
         linewidth=1)
plt.yscale('log')  # Use log scale for current
plt.minorticks_on()
plt.xlabel('Vg (V)')
plt.ylabel('I (A)')
plt.grid(which='both', axis='both', linestyle=':', alpha=0.5)
plt.legend()
plt.ylim([5e-12, 2e-4])
```

```python
region_indices = np.where(Vg < 0.5)
Vg_error = Vg[region_indices]
I_error = I[region_indices]

# Calculate absolute error between fitted and original curves
error = np.abs(I_fit[region_indices] - I_error)

# Calculate relative error (percentage)
relative_error = (error/I_error) * 100

plt.figure(figsize = (10, 6))
plt.plot(Vg_error, relative_error, 'o')

# Add annotations for data points
for i, (x, y) in enumerate((zip(Vg_error, relative_error))):
    if y < 15 and (i+2)%10==0:
    
        # Add the annotation
        current = I[np.where(Vg == x)][0]  # Assuming Vg and I are numpy arrays
        label = "{}mV, {:.2e}A, {:.2}%".format(x * 1000, current, y)
        
        plt.annotate(label,
                    (x, y),  # Point being annotated
                    xytext=(0, 5),  # Offset for text position
                    textcoords="offset points",  # Use offset coordinates
                    ha='center',  # Horizontal alignment
                    rotation=0)  # Rotate text

plt.minorticks_on()
plt.grid(which='both', axis='both', linestyle=':', alpha=0.5)
plt.xlabel('Vg (V)')
plt.ylabel('Relative Error (%)')
plt.title('Relative Error between Subthreshold Channel Current and Exponential Fit', fontsize=14)
plt.ylim([0, 15])  # Show only < 15% error values
```

### Estimate Thermal Voltage, $ V_{th} $
#### Sweep Source Voltage $ V_s $ at a Constant Gate Voltage $ V_g $ in $ I = I_o e^{ \frac{\kappa v_g}{V_{th}} } e^{ -\frac{v_s}{V_{th}} } $

```python
plt.figure(figsize=(10, 6))
plt.plot(Vg, I, label='I-Vg Curve for Vs = 0V')
plt.plot(Vg, I_vs_20m, label='I-Vg for Vs = 20mV')
plt.plot(Vg, I_vs_40m, label='I-Vg for Vs = 40mV')
plt.plot(Vg, I_vs_60m, label='I-Vg for Vs = 60mV')
plt.plot(Vg, I_vs_80m, label='I-Vg for Vs = 80mV')
plt.plot(Vg, I_vs_100m, label='I-Vg for Vs = 100mV')
plt.yscale('log')  # Use log scale for current
plt.xlabel('Vg (V)')
plt.ylabel('I (A)')
plt.minorticks_on()
plt.grid(which='both', axis='both', linestyle=':', alpha=0.5)
plt.legend()
plt.title('Subthreshold Channel Current in Saturation', fontsize=14)
```

```python
# Define the Vg range
imin = int(np.where(Vg == 0.201)[0])
imax = int(np.where(Vg == 0.294)[0])

# Calculate threshold voltage estimates for each gate voltage
all_estimates = []

for i in range(imin, imax+1):
    # Calculate threshold voltage estimates
    Vth_est1 = -0.02/np.log(I_vs_20m[i]/I[i])
    Vth_est2 = -0.02/np.log(I_vs_40m[i]/I_vs_20m[i])
    Vth_est3 = -0.02/np.log(I_vs_60m[i]/I_vs_40m[i])
    Vth_est4 = -0.02/np.log(I_vs_80m[i]/I_vs_60m[i])
    Vth_est5 = -0.02/np.log(I_vs_100m[i]/I_vs_80m[i])

    # Add all estimates to the list
    all_estimates.extend([Vth_est1, Vth_est2, Vth_est3, Vth_est4, Vth_est5])
    
    # Print results formatted to 6 decimal places
    # print(f'Vg = {Vg[i]:.3f}')
    # print(f'Vth estimates from five different values of Vs: {Vth_est1:.6f} {Vth_est2:.6f} {Vth_est3:.6f} {Vth_est4:.6f} {Vth_est5:.6f}\n')
    
# Calculate the final average
final_Vth_estimate = np.mean(all_estimates)
print(f'Final threshold voltage estimate: {final_Vth_estimate:.6f} V')
```

### Estimate $\kappa$ from  Exponential Fit's Slope Expressed as $ \frac{\kappa}{V_{th}} $

```python
kappa = popt[1]*final_Vth_estimate

print(f'kappa estimate: {kappa:.6f}')
```

### Log-Domain Simulation of Single Compartment Neuron with Single Synapse

For multiple compartments and synapses, the system of ODEs is given by:

$$ -\frac{V_{th}}{V_{m_{i}}^*} C \frac{dV_{m_{i}}^{*}}{dt} = \sum_{j}G_{ij}^* (V_{rev_{ij}}^{*} - V_{m_{i}}^{*}) $$

where 

\begin{align}
V_{m_{i}}^{*} &= -V_{th} e^{ -\frac{v_{m_{i}}}{V_{th}} } \\
V_{m_{i}}^{*} &\leq 0 \\
V_{rev_{ij}} &= -V_{th} e^{ -\frac{v_{rev_{ij}}}{V_{th}} } \\
G_{ij}^* &= \frac{1}{V_{th}} I_o \frac{W}{L} e^{ \frac{\kappa v_{g_{ij}}}{V_{th}} } 
\end{align}

$$ V_{m_{i}}^* > \theta_{i}^* = -V_{th} e^{-\frac{\theta_i}{V_{th}}} \Rightarrow \text{Fire} $$


$$ \Rightarrow \Delta G_{ij}^* = \frac{ \kappa I_{o_n} I_{o_p} \frac{W_n}{L_n} \frac{W_p}{L_p}  e^{\frac{-(1 - \kappa)v_{dd}}{V_{th}}} }{ V_{th}^2 C_{syn} } e^{\frac{v_s}{V_{th}}} \Delta t $$

Furthermore, 

$$ G_{ij}^*(t) = G_{ij}^o e^{\frac{-t}{\tau}} $$ 

where 

$$ G_{ij}^0 = \Delta G_{ij}^* $$

```python
def IFATneuron_update(Vm, G, Vrev, Vth, kappa, Ion, Iop, Csyn, Vdd, 
                      Cmem, Vthresh, Vreset, Tau, dt, spike_in, Vs, W, 
                      Trefr, refractory_counter, Vm_max):
    
    if refractory_counter > 0:
        Vm_update = Vreset
        spike_out = 0
        refractory_counter -= dt
    else:
        if Vm < Vthresh:
            Vm_update = Vm + dt * ( 
                (G * (Vrev - Vm)) / (-(Vth / Vm) * Cmem)
            )
            Vm_update = min(Vm_update, Vm_max)  # Clip Vm_update to Vm_max
            spike_out = 0
        else: 
            Vm_update = Vreset
            spike_out = 1
            refractory_counter = Trefr
    
    if spike_in == 1:
        G_update = G + ((kappa * Ion * Iop * np.exp((-(1 - kappa) * Vdd) / Vth)) / (Csyn * (Vth ** 2))) * np.exp(Vs / Vth) * W 
        - dt * (G / Tau)
    else:
        G_update = G - dt * (G / Tau)
        
    return Vm_update, G_update, spike_out, refractory_counter

# Parameters
dt = 1e-6    # timestep for the simulation
Vth = 0.02586    # thermal voltage at room temperature: 25.86 mV
kappa = 0.8
Ion = 4.8577e-12    # 4.8577 pA
Iop = 1.4642e-12    # 1.4642 pA
Csyn = 20e-15    # 20 fF at ~0.6 V moscap gate voltage
Vdd = 1.8    # PMOS body voltage
Cmem = 40e-15    # 40 fF at ~0.5 V moscap gate voltage
theta = 0.3    # neuron membrane threshold voltage
Vthresh = -Vth * np.exp(-theta / Vth)    # neuron membrane threshold voltage in log-domain
vreset = 0.0    # neuron membrane reset voltage
Vreset = -Vth * np.exp(-vreset / Vth)    # neuron membrane reset voltage in log-domain
vrev = 0.5     # synapse reversal potential
Vrev = -Vth * np.exp(-vrev / Vth)    # synapse reversal potential in log-domain
Tau = 1e-3    # time constant for decay of synapse state variable
Vs = 0.6
W = 10e-6
Trefr = 1e-3    # refractory period of the neuron
vm_max = 1.8    # the highest value the membrane potential can reach
Vm_max = -Vth * np.exp(-vm_max / Vth)    # the highest value the membrane potential can reach (in log-domain)

# Initial conditions
vm0 = 0
Vm0 = -Vth * np.exp(-vm0 / Vth)
vg0 = 0
G0 = (1 / Vth) * Ion * np.exp((kappa * vg0) / Vth)

# Time array
T = 10e-3
t = np.arange(0, T + dt, dt)

# Input function (example: regular input spike rate of 1KHz)
input_spike_rate = 1e3
def input_function(t):
    spike_in = np.zeros(len(t))
    for i in range(len(t)):
        if (i+1) % int((1 / dt) / input_spike_rate) == 0:
            spike_in[i] = 1      
    return spike_in

# spike_in = input_function(t)
spike_in = np.zeros(len(t))    # no input spikes

Vm = np.zeros(len(t))
G = np.zeros(len(t))
spike_out = np.zeros(len(t), dtype=int)
refractory_counter = np.zeros(len(t))    # initialise the refractory period counter to all 0s 

Vm[0] = Vm0
G[0] = G0

for i in range(len(t) - 1):
    Vm[i+1], G[i+1], spike_out[i+1], refractory_counter[i+1] = IFATneuron_update(
        Vm[i], G[i], Vrev, Vth, kappa, Ion, Iop, Csyn, Vdd, Cmem, 
        Vthresh, Vreset, Tau, dt, spike_in[i], Vs, W, 
        Trefr, refractory_counter[i], Vm_max)
        
vm = -Vth * np.log(-Vm / Vth)
```

On solving for the analytical expression for the log-domain membrane potential $ V_m $, we get

$$ \frac{-V_{th}}{V_{m_{i}}^*} C_{mem} \frac{dV_{m_{i}}^*}{dt} = G_{ij}^*(t)(V_{rev}^* - V_{m_{i}}^*) $$

where $ G_{ij}^*(t) = G_{ij}^0 e^{\frac{-t}{\tau}} $

$$ -V_{th} C_{mem}\int_{V_{reset}}^{V_{m_{i}}^*} \frac{1}{V_{m_{i}}^*(V_{rev}^* - V_{m_{i}}^*)}\,dV_{m_{i}}^* = G_{ij}^0 \int_0^t e^{\frac{-t}{\tau}}\,dt $$

On multiplying and dividing by $ V_{rev}^* $ and on partial fraction expansion, we get

$$ \frac{-V_{th} C_{mem}}{V_{rev}^*} \int_{V_{reset}}^{V_{m_{i}}^*} \frac{(V_{rev}^* - V_{m_{i}}^*) + (V_{m_{i}}^*)}{V_{m_{i}}^*(V_{rev}^* - V_{m_{i}}^*)}\,dV_{m_{i}}^* = G_{ij}^0 \int_0^t e^{\frac{-t}{\tau}}\,dt $$

$$ \frac{-V_{th} C_{mem}}{V_{rev}^*} \int_{V_{reset}}^{V_{m_{i}}^*} (\frac{1}{V_{m_{i}}^*} + \frac{1}{V_{rev}^* - V_{m_{i}}^*})\,dV_{m_{i}}^* = \tau G_{ij}^0 (1 - e^{\frac{-t}{\tau}}) $$

On integrating, we get

$$ \frac{-C_{mem} V_{th}}{V_{rev}^*}[\ln{(\frac{V_{m_{i}}^*}{V_{reset}})} - \ln{(\frac{V_{rev}^* - V_{m_{i}}^*}{V_{rev}^* - V_{reset}})}] = \tau G_{ij}^0 (1 - e^{\frac{-t}{\tau}}) $$

$$ \large{\Rightarrow \frac{-C_{mem} V_{th}}{V_{rev}^*}\ln{(\frac{V_{m_{i}}^*}{V_{reset}})(\frac{V_{rev}^* - V_{reset}}{V_{rev}^* - V_{m_{i}}^*})} = \tau G_{ij}^0 (1 - e^{\frac{-t}{\tau}})} $$

$$ \Rightarrow V_{m_{i}}^* = \frac{V_{rev}^*}{1 + (\frac{V_{rev}^* - V_{reset}}{V_{reset}}) e^{\frac{-\tau G_{ij}^0 (1 - e^{\frac{-t}{\tau}})}{\frac{-C_{mem} V_{th}}{V_{rev}^*}}}} $$

Percentage error between the estimated and original **steady state values** of the log-domain membrane potential:
$$ \large{\frac{\theta_{\infty_{estimate}} - \theta_{\infty_{original}}}{\theta_{\infty_{original}}}} $$
based on the following expression
$$ \large{\theta_{\infty_{original}} = \frac{V_{rev}^*}{1 + (\frac{V_{rev}^* - V_{reset}}{V_{reset}}) e^{\frac{-\tau * G_{ij}^0}{\frac{-C_{mem}V_{th}}{V_{rev}^*}}}}} $$

```python
theta_inf_est = Vm[len(Vm)-1]
print(f'Estimated theta_inf: {theta_inf_est}')

theta_inf_orig = Vrev / (1 + ((Vrev - Vreset) / Vreset) * np.exp((-Tau * G0) / ((-Cmem * Vth) / Vrev)))
print(f'Predicted theta_inf: {theta_inf_orig}')

theta_inf_error = ((theta_inf_est - theta_inf_orig) / theta_inf_orig) * 100
print(f'Percentage error: {theta_inf_error}%')



G_inf_est = (((-Cmem * Vth) / Vrev) * np.log((theta_inf_est * (Vrev - Vreset)) / (Vreset * (Vrev - theta_inf_est))) 
            / Tau)
print(G_inf_est)

G_inf_orig = G0
print(G_inf_orig)

G_inf_error = ((G_inf_est - G_inf_orig) / G_inf_orig) * 100
print(f'{G_inf_error}%')


# Hypothetical time constant for log-domain membrane potential, Vm*, charging 
tau_hypo = -Tau * np.log(1 - ((-Cmem * Vth)/(Vrev * Tau * G0)) * 
                         np.log(((0.632 * (theta_inf_orig - Vreset) + Vreset) / Vreset) * 
                                ((Vrev - Vreset) / (Vrev - (0.632 * (theta_inf_orig - Vreset) + Vreset)))))
print(f'Hypothetical Time Constant {tau_hypo * 1000} msec')
```

### Log-Domain Analytical Expression for a Single Compartment with Leak Conductance

On solving for the analytical expression for the log-domain membrane potential $ V_m $ with just the and leak conductance and the membrane capacitance, we get

$$ \frac{-V_{th}}{V_{m_{i}}^*} C_{mem} \frac{dV_{m_{i}}^*}{dt} = G_{leak}^* (E_{leak}^* - V_{m_{i}}^*) $$

$$ -V_{th} C_{mem}\int_{V_{reset}}^{V_{m_{i}}^*} \frac{1}{V_{m_{i}}^*(E_{leak}^* - V_{m_{i}}^*)}\,dV_{m_{i}}^* = G_{leak}^* \int_0^t \,dt $$

On multiplying and dividing by $ E_{leak}^* $ and on partial fraction expansion, we get

$$ \frac{-V_{th} C_{mem}}{E_{leak}^*} \int_{V_{reset}}^{V_{m_{i}}^*} \frac{(E_{leak}^* - V_{m_{i}}^*) + (V_{m_{i}}^*)}{V_{m_{i}}^*(E_{leak}^* - V_{m_{i}}^*)}\,dV_{m_{i}}^* = G_{leak}^* \int_0^t \,dt $$

$$ \frac{-V_{th} C_{mem}}{E_{leak}^*} \int_{V_{reset}}^{V_{m_{i}}^*} (\frac{1}{V_{m_{i}}^*} + \frac{1}{E_{leak}^* - V_{m_{i}}^*})\,dV_{m_{i}}^* = G_{leak}^* \,t $$

On integrating, we get

$$ \frac{-C_{mem} V_{th}}{E_{leak}^*}[\ln{(\frac{V_{m_{i}}^*}{V_{reset}})} - \ln{(\frac{E_{leak}^* - V_{m_{i}}^*}{E_{leak}^* - V_{reset}})}] = G_{leak}^* \,t $$

$$ \Rightarrow \frac{-C_{mem} V_{th}}{E_{leak}^*}\ln{(\frac{V_{m_{i}}^*}{V_{reset}})(\frac{E_{leak}^* - V_{reset}}{E_{leak}^* - V_{m_{i}}^*})} = G_{leak}^* \,t $$

$$ \Rightarrow V_{m_{i}}^* = \frac{E_{leak}^*}{1 + (\frac{E_{leak}^* - V_{reset}}{V_{reset}}) e^{\frac{-G_{leak}^* \,t}{\frac{-C_{mem} V_{th}}{E_{leak}^*}}}} $$

$$ \Rightarrow V_{m_{i}}^* = \frac{E_{leak}^*}{1 + (\frac{E_{leak}^* - V_{reset}}{V_{reset}}) e^{\frac{-\,t}{(\frac{C_{mem}}{G_{leak}^*}) (\frac{-V_{th}}{E_{leak}^*})}}} $$

```python
gleak = 0.2
Gleak = (1 / Vth) * Ion * np.exp((kappa * gleak) / Vth)

# Time array
T2 = 100
t2 = np.arange(0, T2 + dt, dt)

eleak = 0.5     # leak potential
Eleak = -Vth * np.exp(-eleak / Vth)    # leak potential in log-domain

Vm_leak = Vrev / (1 + ((Eleak - Vreset) / Vreset) * np.exp((-Gleak * t2) / ((-Cmem * Vth) / Eleak)))

vm_leak = -Vth * np.log(-Vm_leak / Vth)

plt.figure(figsize=(12, 10))

plt.subplot(2, 1, 1)
plt.plot(t2, Vm_leak)
plt.title('Analytical expression for Vm with just leak conductance (log-domain)')
plt.minorticks_on() 
plt.grid(which = 'both', axis = 'x', alpha = 0.5)

plt.subplot(2, 1, 2)
plt.plot(t2, vm_leak)
plt.title('Analytical expression for Vm with just leak conductance')
plt.minorticks_on() 
plt.grid(which = 'both', axis = 'x', alpha = 0.5)
plt.ylabel('Membrane Potential (V)')

plt.tight_layout()
```
